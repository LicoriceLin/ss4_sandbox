```python
from datasets import Dataset
ds=Dataset.load_from_disk('data/reseek_AF_SIWSS-v2')
```

```python
length,plddt=[],[]
from tqdm import tqdm
for i in tqdm(ds):
    l=len(i['pLDDT'])
    if l>0:
        length.append(i)
        plddt.append(sum(i['pLDDT'])/l)
```

```python
import numpy as np
import pandas as pd
df=pd.DataFrame(dict(length=length,plddt=plddt))
```

```python
df['length'][0]
```

```python
df['length']=df['length'].apply(lambda x:len(x['pLDDT']))
```

```python
df.to_pickle('data/AF_SWISS_mufactorized_debug/distribution.pkl')
```
